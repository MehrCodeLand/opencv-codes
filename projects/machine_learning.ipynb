```python
import pandas as pd
import numpy as np 
import seaborn as sns 
from sklearn import datasets
```

```python


df = pd.read_csv('../Datas/Advertising.csv')
ds = datasets.load_diabetes()
```

```python
ds.feature_names
```

```python
X = ds.data
Y = ds.target
```

```python
X.shape , Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.3 , random_state=42)
```

```python
X_train.shape , Y_train.shape
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
mean_squared_error(Y_test,y_pred)
```

```python
mean_absolute_error(Y_test,y_pred)
```

```python
r2_score(Y_test, y_pred)
```

```python
sns.scatterplot(y=Y_test , x=y_pred )
```

```python
X = df.drop("Sales" , axis = 1 )
X = X.drop(columns= ["Unnamed: 0"], axis = 1 )
X.head()
```

```python
y = df["Sales"]
```

```python
 from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_convert = PolynomialFeatures(degree=2 , include_bias=False)
```

```python
polynomial_convert.fit(X)
```

```python
poly_fetures = polynomial_convert.transform(X)
poly_fetures
```

```python
poly_fetures.shape
```

```python
X.iloc[0]
```

```python
poly_fetures[0]
```

```python
X_train , X_test , y_train , y_test = train_test_split(
    poly_fetures , 
    y ,
    test_size=0.3,
    random_state=101
)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept= True)
```

```python
model.fit(X_train , y_train)
```

```python
test_predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test , test_predictions)
```
